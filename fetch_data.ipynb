```python
import pandas as pd

import numpy as np

import plotnine as p9

import fetch_data

import importlib
importlib.reload(fetch_data)

import requests
from bs4 import BeautifulSoup
```

```python
sites = fetch_data.get_colorado_sites()

sites
```

```python
all_sites_data = []
for row_num, row in sites.iterrows():
    station = row['station']

    raw_data = fetch_data.download_site_data(station)
    df = fetch_data.parse_snow_data(raw_data, station)
    all_sites_data.append(df)

all_sites_df = pd.concat(all_sites_data, ignore_index=True)\
    .merge(sites, on='station', how='left')

all_sites_df
```

```python
all_sites_df_pivot = fetch_data.pivot_to_long_format(all_sites_df)

all_sites_df_pivot
```

```python
larimer_county = all_sites_df_pivot.query("county == 'Larimer'")

larimer_county['site_name'].value_counts()
```
